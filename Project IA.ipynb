### Transform data using Label Encouder

```python
from sklearn import preprocessing as sk
encouder = sk.LabelEncoder()
df["Project Description"]=encouder.fit_transform(df["Project Description"])
df["Project Phase Name"]=encouder.fit_transform(df["Project Phase Name"])
df["Project_Type"]=encouder.fit_transform(df["Project_Type"])
df
```

```python
df.drop('DSF Number(s)',axis='columns', inplace=True) 
df.drop('Project Geographic District',axis='columns', inplace=True) 
df.drop('Project Building Identifier',axis='columns', inplace=True) 
df.drop('Project School Name',axis='columns', inplace=True) 
df.drop('Project Status Name',axis='columns', inplace=True) 

```

### Changer the dates to  dateTime object


```python
from datetime import datetime


dateFormatter = "%m/%d/%Y"
date1=[]
for date in df["Project Phase Actual Start Date"]:

    
    date1.append(datetime.strptime(date, dateFormatter))
date1
df["date_start_project "]=date1
df
```

```python
date2=[]
for date in df["Project Phase Actual End Date"]:

    
    date2.append(datetime.strptime(date, dateFormatter))
df["actual_date_end"]=date2
df
```

```python
#df[df["Project Budget Amount"]=='FTK']='0'
df["Project Budget Amount"].value_counts()
#df["Project Budget Amount"].astype('float')
```

```python
df["Project Budget Amount"]=df["Project Budget Amount"].apply(lambda x: 0 if x=='FTK' or x=='DIIP' or x=='DOES' or x=='IEH' else x)

#df[df["Project Budget Amount"]=='0']=df["Final Estimate of Actual Costs Through End of Phase Amount"]
df["Project Budget Amount"].value_counts()
df["Project Phase Planned End Date"].value_counts()
```

```python
for i, row in df.iterrows():  # i: dataframe index; row: chaque ligne (row) au format series
    if df.iloc[i,4]=="DOES" or df.iloc[i,4]=="IEH" or df.iloc[i,4]=="DIIR" or df.iloc[i,4]=="DIIP" or  df.iloc[i,4]=="FTK" or df.iloc[i,4]=="DOER" or df.iloc[i,4]=="EMER" :
        df.iloc[i,4]=df.iloc[i,5]
df
df["Project Phase Planned End Date"].value_counts()
```

```python
date3=[]
for date in df["Project Phase Planned End Date"]:

    
    date3.append(datetime.strptime(date, dateFormatter))
df["planned_date"]=date3

df["Project Budget Amount"].value_counts()
```

### Calculate the real and estimation duration and create new features to use it in prediction 

```python
listeDureeEstimee=[]
for i, row in df.iterrows():  # i: dataframe index; row: chaque ligne (row) au format series
    x=df.iloc[i,11]-df.iloc[i,9]
    listeDureeEstimee.append(x)
df["duree_estimee"]=listeDureeEstimee
df
```

```python
listeDureeReelle=[]
for i, row in df.iterrows():  # i: dataframe index; row: chaque ligne (row) au format series
    x=df.iloc[i,10]-df.iloc[i,9]
    listeDureeReelle.append(x)
df["duree_reelle"]=listeDureeReelle
df
df.info()
```

### Drawing the correlation map to understand the correlation between features 

```python
import matplotlib.pyplot as plt
import seaborn as sns
def plot_correlation_map( df ):

    corr = df.corr()

    s , ax = plt.subplots( figsize =( 10 , 10 ) )

    cmap = sns.diverging_palette( 220 , 10 , as_cmap = True )

    s = sns.heatmap(

        corr, 

        cmap = cmap,

        square=True, 

        cbar_kws={ 'shrink' : .9 }, 

        ax=ax, 

        annot = True, 

        annot_kws = { 'fontsize' : 12 }

        )
```

```python
plot_correlation_map( df_copie )
```

###  Correction model for time estimation using linear regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

X=df_copie[["duree_estimee_jour"]]
y=df_copie["duree_reelle_jour"]
X_train,X_test,Y_train,Y_test= train_test_split(X,y,test_size=0.20,random_state=30)
model=LinearRegression()   #build linear regression model
model.fit(X_train,Y_train)  #fitting the training data
```

### plot the linear regression for this model 

```python
plt.plot(X,y,'o')
plt.title("correction duration estimate")
plt.ylabel("duree_reelle_jour")
plt.xlabel("duree_estimee_jour")
plt.plot(X,model.predict(X),'o',color="k")
plt.show()
```

### Measure the performance of linear regression of this model using the testing set

```python
from sklearn.metrics import mean_squared_error
from sklearn import metrics
predicted=model.predict(X_test)
print("MSE", mean_squared_error(Y_test,predicted))
print("R squared", metrics.r2_score(Y_test,predicted))
```

###  Correction model for the budget estimate using linear regression

```python
X2=df_copie[["Project_budget_amount"]]
y2=df_copie["Total Phase Actual Spending Amount"]
X2_train,X2_test,Y2_train,Y2_test= train_test_split(X2,y2,test_size=0.20,random_state=30)
model2=LinearRegression()   #build linear regression model
model2.fit(X2_train,Y2_train)  #fitting the training data
```

```python
plt.plot(X2,y2,'o')
plt.title("budget correction")
plt.ylabel("Total Phase Actual Spending Amount")
plt.xlabel("Project_budget_amount")
plt.plot(X2,model2.predict(X2),'o',color="k")
plt.show()
```

```python
predicted=model2.predict(X2_test)
print("MSE", mean_squared_error(Y2_test,predicted))
print("R squared", metrics.r2_score(Y2_test,predicted))
```
